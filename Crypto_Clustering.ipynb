```python
# Import required libraries and dependencies
import pandas as pd
import os
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
df_market_data.columns
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_numeric_columns = df_market_data.select_dtypes(include='number').columns
crypto_scaled_data = StandardScaler().fit_transform(df_market_data[crypto_numeric_columns])
```

```python
# Create a DataFrame with the scaled data
df_crypto_scaled = pd.DataFrame(crypto_scaled_data, columns=crypto_numeric_columns, index=df_market_data.index)

# Copy the crypto names from the original DataFrame ---- WHY?
# Set the coin_id column as index --- WHY?


# Display the scaled DataFrame
df_crypto_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):  
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_crypto_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)
```

```python
print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": list(range(1, 12)),  # k-values from 1 to 11
    "inertia": inertia       # List of inertia values computed earlier
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
print(df_elbow)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow = df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800,
    height=400
)

elbow
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for "k" is "4".

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans = KMeans(n_clusters=4, random_state=10)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(df_crypto_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans.predict(df_crypto_scaled)

# Print the resulting array of cluster values.
print(f'Clusters: {clusters}')
```

```python
# Create a copy of the scaled DataFrame
df_crypto_scaled_copy = df_crypto_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_crypto_scaled_copy['cluster'] = clusters

# Display the copy of the scaled DataFrame
df_crypto_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
scatter_plot = df_crypto_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h", # x="price_change_percentage_24h" 
    y="price_change_percentage_7d", # `y="price_change_percentage_7d"`
    xlabel="24h Price Change (%)",
    ylabel="7d Price Change (%)",
    c="cluster",  # Color by K-Means cluster
    cmap=["pink", "red", "orange", "yellow"],  # Color code 
    hover_cols=["coin_id"],  # Show the crypto name on hover
    title="Cryptocurrency Clusters Based on Price Changes",
    width=800,
    height=600
)

# Display the scatter plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_crypto_scaled)

# View the scaled PCA data
pca_data
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

print("Explained variance:", explained_variance)

print("Total explained variance:", explained_variance.sum())
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Explained Variance: PC1 = 0.3719856, PC2 = 0.34700813, PC3 = 0.17603793, Total explained variance = 0.895031657030984

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original scaled DataFrame
df_pca['coin_id'] = df_market_data.index

# Set the coin_id column as index
df_pca.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
df_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):  
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_pca)
# 3. Append the model.inertia_ to the inertia list
    inertia_pca.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {
    "k": list(range(1, 12)),  # k-values from 1 to 11
    "inertia": inertia_pca       # List of inertia values computed earlier
}

# Create a DataFrame with the data to plot the Elbow curve
pca_df_elbow = pd.DataFrame(pca_elbow_data)
print(pca_df_elbow)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow = pca_df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800,
    height=400
)

pca_elbow 
```

  * **Answer:** The best value for "k" is "4".

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans = KMeans(n_clusters=4, random_state=10)
```

```python
# Fit the K-Means model using the PCA data
kmeans.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_clusters = kmeans.predict(df_pca)

# Print the resulting array of cluster values.
print(f'Clusters: {pca_clusters}')
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_copy = df_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters

df_pca_copy['cluster'] = pca_clusters

# Display the copy of the scaled PCA DataFrame
df_pca_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
pca_scatter_plot = df_pca_copy.hvplot.scatter(
    x="PC1", # x="PC1" 
    y="PC2", # `y="PC2"`
    xlabel="PC1",
    ylabel="PC2",
    c="cluster",  # Color by K-Means cluster
    cmap=["blue", "green", "purple", "brown"],  # Color code 
    hover_cols=["coin_id"],  # Show the crypto name on hover
    title="Cryptocurrency Clusters Based on PCA",
    width=800,
    height=600
)

# Display the scatter plot
pca_scatter_plot
```
